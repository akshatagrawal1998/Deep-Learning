```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights = 'imagenet') # we need weihts which were obtained after training ImageNet dataset
```

Add images to test in colab like I added 2 images for testing purpose

```python
img_path = '/content/car.jpg'
img = image.load_img(img_path, target_size=(224,224)) # load the image
x = image.img_to_array(img) # converted image to array
x = np.expand_dims(x, axis=0) # convert image to batch
x = preprocess_input(x) # preprocess the image
```

```python
pred = model.predict(x)
```

```python
print("Predictions : ", decode_predictions(pred, top=3)[0])
```

We see perfect results, it is telling that the input image is a car
